```python
import pandas as pd
```

```python
df=pd.read_csv(r"data-final.csv",sep="\t")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df
```

```python
x=df.iloc[:6000,0:50]
x
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5)

kmeans = kmeans.fit(x)
```

```python
kmeans.labels_
```

```python
predictions = kmeans.labels_
predictions=pd.DataFrame(predictions)
```

```python
from sklearn.decomposition import PCA as sklearnPCA
sklrn_pca = sklearnPCA(n_components=2)
Y_sklrn = sklrn_pca.fit_transform(x)
Y_sklrn=pd.DataFrame(Y_sklrn)
```

```python
Y_sklrn
```

```python
Y_sklrn['Clusters'] = predictions
```

```python
Y_sklrn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(data=Y_sklrn,x=Y_sklrn[0],y=Y_sklrn[1],hue='Clusters')
plt.xlabel("PrincipalComponent1")
plt.ylabel("PrincipalComponent2")
plt.title("PCA")
```

```python
col_list = list(x)
extro = col_list[0:10]
neu = col_list[10:20]
agr = col_list[20:30]
csn = col_list[30:40]
op = col_list[40:50]

dsum = pd.DataFrame()
dsum['extroversion'] = x[extro].sum(axis=1)/10
dsum['neurotic'] = x[neu].sum(axis=1)/10
dsum['agreeable'] = x[agr].sum(axis=1)/10
dsum['conscientious'] = x[csn].sum(axis=1)/10
dsum['open'] = x[op].sum(axis=1)/10
dsum['clusters'] = predictions
dsum.groupby('clusters').mean()
```

```python
clusters = dsum.groupby('clusters').mean()
for i in range(0, 5):
    plt.bar(clusters.columns, clusters.iloc[i, :],color='lightblue')
    plt.plot(clusters.columns, clusters.iloc[i, :], color='red')
    plt.title('Cluster ' + str(i))
    plt.show()
```
